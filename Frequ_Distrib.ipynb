**Importando as bibliotecas**

```python
import pandas as pd
import numpy as np
import plotly.express as px
from collections import Counter
from pickle import TRUE
```

**Carregando o dataset**

```python
df=pd.read_csv('/content/sample_data/adult.csv')
df.head()
```

**Selecionando somente as colunas que vamos usar**

```python
df = pd.read_csv('/content/sample_data/adult.csv', usecols=['age', 'marital.status', 'race', 'sex'])
df.head()
```

**Qual é  o números de pessoas por idade?**

*Calcular da frequência absoluta*

```python
#Criando um dicionário de dados para realizar a contagem de pessoas por idade
idade = df['age']
my_dict= Counter(idade)
my_dict
print(my_dict)
```

**Transformando em um dataframe**

```python
df1 = pd.DataFrame(list(my_dict.items()),columns = ['idade','freq_abs'])  
df1.sort_values('idade', axis=0, ascending=True, inplace=True)
df1
```

**Quantas pessoas de 17 a 20 anos existem?**

*Calcular da frequência absoluta acumulada*

```python
df1['freq_abs_acum'] = df1['freq_abs'].cumsum()
df1.head(4)
```

**Qual a porcentagem de pessoas por idade existem?**

*Calcular da frequência absoluta relativa*

```python
df1['freq_abs_rel %'] = df1['freq_abs'] / df1['freq_abs'].sum()*100
df1
#Existem 1.6% de pessoas com 18 anos
```

**Qual a porcentagem de pessoas existem até 21 anos?**

*Calcular da frequência absoluta acumulada relativa*

```python
df1['freq_abs_acum_rel %'] = df1['freq_abs_rel %'].cumsum()
df1.head(5)
#Existem 9.61% de pessoas até 21 anos
```
